```python
import os

folder = "images"
images = sorted([
    f for f in os.listdir(folder)
    if f.endswith((".png", ".jpg", ".jpeg", ".pdf"))
])

for i in range(0, len(images), 2):
    print("\\begin{figure}[htbp]")
    print("    \\centering")
    for j in range(2):
        if i + j < len(images):
            print(f"""    \\begin{{minipage}}[t]{{0.48\\textwidth}}
        \\centering
        \\includegraphics[width=\\linewidth]{{{folder}/{images[i + j]}}}
        \\caption*{{{images[i + j]}}}
    \\end{{minipage}}%""")
        if j == 0:
            print("    \\hfill")
    print("\\end{figure}")
    print("\\newpage")
```
